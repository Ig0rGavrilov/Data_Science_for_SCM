```python
import numpy as np
import pandas as pd

array = [28, 19, 18, 13, 19, 16, 19, 18, 13, 100, 16, 11, 18, 15, 13, 15, 13, 11, 13, 10, 12]



higher_limit = np.percentile(array, 99)
lower_limit = np.percentile(array, 1)

clean_array = np.clip(array, a_min= lower_limit, a_max=higher_limit)

df = pd.DataFrame({'Before cleaning': array, 'After cleaning': clean_array, 'Lower bound':lower_limit, 'Higher bound':higher_limit})


df.plot( title='Winsorization')

```
